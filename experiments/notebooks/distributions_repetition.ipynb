```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from cir.benchmarks.cir_utils import get_per_entity_prob, get_pmf
```

# Repetition

```python
n_classes = 20
dist = {"n_entities": n_classes, "dist_type": "zipf", "exponent": 0.7}
probs = get_per_entity_prob(**dist)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 5))
data = {"Class": list(range(n_classes)), "Probability": probs}
sns.barplot(data=data, x="Class", y="Probability", ax=ax)
ax.set_xticks(np.arange(0, n_classes, 2))
ax.set_xlabel("Class", fontsize=20)
ax.set_ylabel("Probability of Repetition", fontsize=20)
ax.set_ylim(0, 1)
plt.tight_layout()
plt.savefig(f"./repetition.pdf")
```
